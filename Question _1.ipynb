```python
def resizeNN(image, resize_width, resize_height):
    """
    :params
        image: shape -> (width, height, chs)
        resize_width: The resize width dimension. 
        resize_height: The resize height dimension. 
    
    :returns
        array of shape -> (resized_width, resized_height, chs)
    """
    
    ow, oh, ch = image.shape
    
    r   = image[:, :, 0]
    g = image[:, :, 1]
    b  = image[:, :, 2]
    
    resized_image = np.zeros((resize_width, resize_height, ch), dtype=np.uint8)
    
    x_scale = ow/resize_width
    y_scale = oh/resize_height
    
    resize_idx = np.zeros((resize_width, resize_height))
    resize_index_x = np.ceil(np.arange(0, ow, x_scale)).astype(int)
    resize_index_y = np.ceil(np.arange(0, oh, y_scale)).astype(int)
    resize_index_x[np.where(resize_index_x == ow)]  -= 1
    resize_index_y[np.where(resize_index_y == oh)] -= 1

    resized_image[:, :, 0] = r[resize_index_x, :][:, resize_index_y]
    resized_image[:, :, 1] = g[resize_index_x, :][:, resize_index_y]
    resized_image[:, :, 2] = b[resize_index_x, :][:, resize_index_y]
    
    return resized_image
```

```python
resized_by_func=resizeNN(img1, 222,296)
```

```python
cv2.imshow("Func resized Image", resized_by_func)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def RMSError(a1, a2):
    assert a1.shape == a2.shape
    rmse =np.sqrt( np.mean((a1 - a2) ** 2))
    return rmse
```
