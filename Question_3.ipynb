```python
def rotate(image, degree):

    in_radians = math.radians(degree)

    rotated_img = np.uint8(np.zeros(image.shape))
    height = rotated_img.shape[0]
    width  = rotated_img.shape[1]

    center_x = width//2
    center_y = height//2

    for i in range(rotated_img.shape[0]):

        for j in range(rotated_img.shape[1]):
            # applying the formula of image rotation
            x= (i-center_x)*math.cos(in_radians)+(j-center_y)*math.sin(in_radians)
            y= -(i-center_x)*math.sin(in_radians)+(j-center_y)*math.cos(in_radians)

            x=round(x)+center_x 
            y=round(y)+center_y 

            if (x>=0 and y>=0 and x<image.shape[0] and  y<image.shape[1]):
                rotated_img[i,j,:] = image[x,y,:]

    return rotated_img
```

```python
img1=cv2.imread('img1.jpg')
rotate15user = rotate(img1,15)

cv2.imshow("rotated image",rotate15user)
cv2.waitKey(0)
```

```python
rotate45user = rotate(img1,15)

cv2.imshow("rotated image",rotate45user)
cv2.waitKey(0)
```

```python
def RMSError(a1, a2):
    assert a1.shape == a2.shape
    rmse =np.sqrt( np.mean((a1 - a2) ** 2))
    return rmse
```
